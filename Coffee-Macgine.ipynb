```python
class CoffeeMachine:
  import sys

  #water,milk,coffee,cups,money=400,540,120,9,550
  

  def __init__(self):
    self.water = 400
    self.milk = 540
    self.coffee = 120 
    self.cups = 9
    self.money = 550
    self.price=0
    self.coffee_type_dict = {"1":[250,0,16,4],"2":[350,75,20,7],"3":[200,100,12,6]}
    self.coffee_ingrid =["water","milk","coffee beans","cups"] 

  # make a dict with values and pull it from there for the method
  

  def choise(self):
    while True:
      print()
      action=input("Write action (buy, fill, take, remaining, exit):\n")
      if action == "remaining":
        self.machine_supply()
      if action == "buy":
        self.buy()
      elif action == "fill":
        self.fill()
      elif action == "take":
        self.take()
      elif action == "exit":
        self.exit_menu()
      

  def buy(self):
    print()
    coffee_choise = input("What do you want to buy? 1 - espresso, 2 - latte, 3 - cappuccino, back - to main menu:\n")
    if coffee_choise == "back":
      choise()
    else :
      if self.enought_supply(coffee_choise) == "Yes":
        global water,milk,coffee,price,money,cups
        coffee_values = [val for i,val in self.coffee_type_dict.items() if coffee_choise == i]
        self.water -= coffee_values[0][0]
        self.milk -= coffee_values[0][1]
        self.coffee -= coffee_values[0][2]
        self.price = coffee_values[0][3]
        self.cups-=1
        self.money += self.price
        print("I have enough resources, making you a coffee!")
      else : 
        print(f"Sorry, not enough {self.coffee_ingrid[int(self.enought_supply(coffee_choise))]}!")

  def fill(self):
    global water,milk,coffee,cups
    self.water+=int(input("Write how many ml of water you want to add:\n"))
    self.milk+=int(input("Write how many ml of milk you want to add:\n"))
    self.coffee+=int(input("Write how many grams of coffee beans you want to add:\n"))
    self.cups+=int(input("Write how many disposable coffee cups you want to add:\n"))

  def take(self):
    global money
    print(f"I gave you {self.money}")
    self.money=0
    
  def machine_supply(self):
    print(f"""
    The coffee machine has:
    {self.water} of water
    {self.milk} of milk
    {self.coffee} of coffee beans
    {self.cups} of disposable cups
    ${self.money} of money
    """)

  def enought_supply(self, coffee_choise):
    global water,milk,coffee,cups,money
    coffee_values = [val for i,val in self.coffee_type_dict.items() if coffee_choise == i]
    local_list = [self.water, self.milk, self.coffee, self.cups]
    result = [f"{k}"for k,value_1 in enumerate(coffee_values[0]) if local_list[k] < value_1]
    if len(result)>0:
      result=result[0]
    else:
      result="Yes"
    return result
    
  def exit_menu():
    sys.exit()

```
