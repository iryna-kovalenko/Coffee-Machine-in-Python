```python
import sys

water,milk,coffee,cups,money=400,540,120,9,550
price=0

# make a dict with values and pull it from there for the method
coffee_type_dict = {"1":[250,0,16,4],"2":[350,75,20,7],"3":[200,100,12,6]}
coffee_ingrid =["water","milk","coffee beans","cups"]

def choise():
  while True:
    print()
    action=input("Write action (buy, fill, take, remaining, exit):\n")
    if action == "remaining":
      machine_supply()
    if action == "buy":
      buy()
    elif action == "fill":
      fill()
    elif action == "take":
      take()
    elif action == "exit":
      exit_menu()
    

def buy():
  print()
  coffee_choise = input("What do you want to buy? 1 - espresso, 2 - latte, 3 - cappuccino, back - to main menu:\n")
  if coffee_choise == "back":
    choise()
  else :
    if enought_supply(coffee_choise) == "Yes":
      global water,milk,coffee,price,money,cups
      coffee_values = [val for i,val in coffee_type_dict.items() if coffee_choise == i]
      water -= coffee_values[0][0]
      milk -= coffee_values[0][1]
      coffee -= coffee_values[0][2]
      price = coffee_values[0][3]
      cups-=1
      money += price
      print("I have enough resources, making you a coffee!")
    else : 
      print(f"Sorry, not enough {coffee_ingrid[int(enought_supply(coffee_choise))]}!")

def fill():
  global water,milk,coffee,cups
  water+=int(input("Write how many ml of water you want to add:\n"))
  milk+=int(input("Write how many ml of milk you want to add:\n"))
  coffee+=int(input("Write how many grams of coffee beans you want to add:\n"))
  cups+=int(input("Write how many disposable coffee cups you want to add:\n"))

def take():
  global money
  print(f"I gave you {money}")
  money=0
  
def machine_supply():
  print(f"""
  The coffee machine has:
  {water} of water
  {milk} of milk
  {coffee} of coffee beans
  {cups} of disposable cups
  ${money} of money
  """)

def enought_supply(coffee_choise):
  global water,milk,coffee,cups,money
  coffee_values = [val for i,val in coffee_type_dict.items() if coffee_choise == i]
  local_list = [water,milk,coffee,cups]
  result = [f"{k}"for k,value_1 in enumerate(coffee_values[0]) if local_list[k] < value_1]
  if len(result)>0:
    result=result[0]
  else:
    result="Yes"
  return result
  
def exit_menu():
  sys.exit()

choise()
```
